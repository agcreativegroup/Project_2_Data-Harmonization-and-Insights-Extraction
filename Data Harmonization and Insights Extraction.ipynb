# missing value analysis and outlier detection in final_dataset

```python
print("\n--- Missing Value Analysis ---")
missing_percentage = final_dataset.isnull().mean() * 100
print(missing_percentage[missing_percentage > 0])
```

# Handle missing values

```python
warnings.filterwarnings("ignore")
for column in final_dataset.columns:
    if final_dataset[column].isnull().sum() > 0:
        if final_dataset[column].dtype == 'object':  # Categorical variables
            final_dataset[column] = final_dataset[column].fillna(method='bfill').fillna(method='bfill')
        else:  # Numerical variables
            final_dataset[column] = final_dataset[column].interpolate(method='linear').fillna(method='bfill').fillna(method='ffill')
```

```python
# Reinspect final dataset
print("\n--- Final Dataset Info ---")
final_dataset.info()
```

```python
print("\n--- Final Dataset Head ---")
print(final_dataset.head())
```

# Outlier Detection and Handling

```python
print("\n--- Outlier Detection ---")
for column in final_dataset.select_dtypes(include=['float64', 'int64']).columns:
    q1 = final_dataset[column].quantile(0.25)
    q3 = final_dataset[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = final_dataset[(final_dataset[column] < lower_bound) | (final_dataset[column] > upper_bound)]
    print(f"Column: {column}\n Outliers Detected: {len(outliers)}\n")
```

```python
# Apply capping
final_dataset[column] = np.where(final_dataset[column] < lower_bound, lower_bound, final_dataset[column])
final_dataset[column] = np.where(final_dataset[column] > upper_bound, upper_bound, final_dataset[column])
```

```python
# Exploratory Data Analysis (EDA)
print("\n--- Exploratory Data Analysis ---")
numerical_columns = final_dataset.select_dtypes(include=['float64', 'int64']).columns

numerical_columns
```

```python
# Plot histograms
columns_to_histograms = ['instant','temp', 'hum', 'windspeed','cnt','weekday','atemp','casual','registered', 'hr', 'dteday']
n_cols = 3  
n_rows = (len(columns_to_histograms) + n_cols - 1) // n_cols  

fig, axes = plt.subplots(n_rows, n_cols, figsize=(21, 6 * n_rows))  # subplots

axes = axes.flatten()

# using Loop create a histograms 
for i, column in enumerate(columns_to_histograms):
    sns.histplot(final_dataset[column], kde=True, bins=30, color='red', ax=axes[i])
    axes[i].set_title(f"Histogram for {column}")
    axes[i].set_xlabel(column)
    axes[i].set_ylabel("Frequency")
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()  
plt.show()
```

```python
# Plot boxplots
columns_to_boxplot = ['temp', 'hum', 'windspeed','cnt','weekday','atemp','casual','registered', 'hr','instant', 'dteday']
n_cols = 4  
n_rows = (len(columns_to_boxplot) + n_cols - 1) // n_cols  

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows)) 
axes = axes.flatten()

# using Loop create a boxplots 
for i, column in enumerate(columns_to_boxplot):
    sns.boxplot(x=final_dataset[column], color='green', ax=axes[i])
    axes[i].set_title(f"Boxplot for {column}")
    axes[i].set_xlabel(column)

for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()  
plt.show()
```

# Week 3: Advanced Statistical Analysis

```python
# Analyze Skewness
print("\n--- Skewness Analysis ---")
skewness = final_dataset[numerical_columns].skew()
print(skewness)
```

```python
# Applying log transformations to normalize distributions
for column in numerical_columns:
    if skewness[column] > 1 or skewness[column] < -1:
        print(f"Applying log transformation to {column}")
        final_dataset[column] = np.log1p(final_dataset[column])
```
